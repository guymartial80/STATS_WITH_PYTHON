```python
import numpy as np
import matplotlib.pyplot as plt
```

"""
La fonction linspace crée un vecteur avec des points des nombres espacés linéairement. Si on considère np.linspace(a, b, c):
"a" est la limite inférieure des points générés par la fonction ; 
"b" est la limite supérieure des points générés par la fonction ; 
"b/(c-1)" cette valeur sera continuellement additionnée par celle précédemment obtenue jusqu'à atteindre la limite supérieure "b".

Exemple :
np.linspace(0, 1, 5)

a = 0 ;
b = 1 ;
c = 5 et b/(c-1) = 0.25

Le vecteur généré par np.linspace est array([0.  , 0.25, 0.5 , 0.75, 1.  ])

```python
x = np.linspace(0, 10, 100)  
y = np.sin(x)

# Création du graphique
plt.plot(x, y) # Renseigner les coordonnées en abscisse et en ordonnée
plt.show()
```

```python
x = np.linspace(0, 10, 100)  
y = np.sin(x)

# Création du graphique en renseignant les coordonnées
plt.plot(x, y) 

# Titre du graphique
plt.title('Graphique sin(x)')

# Titre axe des abscisses
plt.xlabel('x')

# Titre axe des ordonnées
plt.ylabel('sin(x)')

# Affichage de la grille
plt.grid(True)

# Sauvegarde du graphique
plt.savefig('graphique_sin.png')  

# Affichage du graphique
plt.show()
```

```python
x = np.linspace(0, 10, 100)  

# Graphiques sur deux lignes, une colonne
fig, axs = plt.subplots(2, 1)

# Premier graphique en rouge ('r')
axs[0].plot(x, np.sin(x), 'r')
axs[0].set_title('Graphique sin(x)')

# Deuxième graphique en vert ('g')
axs[1].plot(x, np.cos(x), 'g')
axs[1].set_title('Graphique cos(x)')

# Espace entre graphiques
plt.tight_layout()

# Affichage des graphiques
plt.show()
```

```python
import pandas as pd
import seaborn as sns

# Chargement des données
tips = pd.read_csv('tips.csv')

tips.head()
```

```python
# Création d'un scatter plot
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter plot des pourboires')
plt.show()
```

```python
# Création d'un box plot
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box plot des notes par jour')
plt.show()
```

```python
# Box plot avec Seaborn
sns.boxplot(x='day', y='total_bill', hue='day', data=tips, palette='pastel')
plt.title('Box plot personnalisé des notes par jour')
plt.show()
```
